### Writing Functions to get the desired columns

```python
#This function will gives the value in the form of new column called 'Weight slab as per X (KG)'
def sab_data1(obj):
    L = []
    for i in obj:
        x = np.ceil(i)
        y = x - 0.5
        if i > y:
            L.append(x)
        else:
            L.append(y)
    return L
```

```python
#Calling the function and assigning to the new column
new1['Weight slab as per X (KG)'] = sab_data1(new1['Total weight as per X (KG)'])
```

### Writing the functions

```python
#This function will gives the value in the form of new column called "Company_Fixed_Values"
def fixed_values(obj):
    dictionary = {'Forward charges_A':29.5,
                  'Forward charges_B':33,
                  'Forward charges_C':40.1,
                  'Forward charges_D':45.4,
                  'Forward charges_E':56.6,
                  'Forward and RTO charges_A':43.1,
                  'Forward and RTO charges_B':53.5,
                  'Forward and RTO charges_C':72,
                  'Forward and RTO charges_D':86.7,
                  'Forward and RTO charges_E':407.3}
    L = []
    for i in obj:
        if i in dictionary:
            L.append(dictionary[i])
    return L
```

```python
#Calling the function and assigning to the new column
new1["Company_Fixed_Values"] = fixed_values(new1["fwd_rto_zone"])
```

```python
new1[["fwd_rto_zone","Company_Fixed_Values"]].head()
```

```python
#combining the columns and assigning to the new variable
new1["fwd_rto_zone_additional"] = new1["fwd_rto_zone"] +"_" + "Additional"
```

```python
new1["fwd_rto_zone_additional"]
```

```python
#This function will gives the value in the form of new column called "Company_Additional_Values"
def additional_values(obj):
    dictionary = {'Forward charges_A_Additional':23.6,
                  'Forward charges_B_Additional':28.3,
                  'Forward charges_C_Additional':38.9,
                  'Forward charges_D_Additional':44.8,
                  'Forward charges_E_Additional':55.5,
                  'Forward and RTO charges_A_Additional':47.2,
                  'Forward and RTO charges_B_Additional':56.6,
                  'Forward and RTO charges_c_Additional':77.8,
                  'Forward and RTO charges_D_Additional':89.6,
                  'Forward and RTO charges_E_Additional':111}
    L = []
    for i in obj:
        if i in dictionary:
            L.append(dictionary[i])
    return L
```

```python
##Calling the function and assigning to the new column
new1["Company_Additional_Values"] = additional_values(new1["fwd_rto_zone_additional"])
```

```python
new1[["fwd_rto_zone_additional","Company_Additional_Values"]].head()
```

```python
# Finding out the "Expected Charge as per X (Rs.)"
new1["Expected Charge as per X (Rs.)"] = new1["Company_Fixed_Values"] + (new1['Additional_Difference']*new1["Company_Additional_Values"])
```

```python
#Finding out the "Difference Between Expected Charges and Billed Charges (Rs.)"
new1["Difference Between Expected Charges and Billed Charges (Rs.)"] = new1["Expected Charge as per X (Rs.)"] - new1["Charges Billed by Courier Company (Rs.)"]
```
